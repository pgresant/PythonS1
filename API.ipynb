```python
import pandas as pd
import requests
import py7zr
import tempfile
import os

def read_csv_from_7z_url(url, encoding='latin1', sep=';'):
    # 1. CrÃ©er un fichier temporaire pour le .7z
    with tempfile.TemporaryDirectory() as tmpdir:
        archive_path = os.path.join(tmpdir, "data.7z")
        extracted_path = os.path.join(tmpdir, "extracted")

        # 2. TÃ©lÃ©charger l'archive
        r = requests.get(url, stream=True)
        r.raise_for_status()
        with open(archive_path, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)

        # 3. Extraire l'archive
        with py7zr.SevenZipFile(archive_path, mode='r') as z:
            z.extractall(path=extracted_path)

        # 4. Trouver le premier fichier CSV
        for root, _, files in os.walk(extracted_path):
            for file in files:
                if file.endswith(".csv"):
                    csv_file_path = os.path.join(root, file)
                    return pd.read_csv(csv_file_path, encoding=encoding, sep=sep)

    raise FileNotFoundError("Aucun fichier CSV trouvÃ© dans l'archive.")

# Exemple d'utilisation :
url_SAE_2017 = "https://drees2-sgsocialgouv.opendatasoft.com/api/datasets/1.0/708_bases-statistiques-sae/attachments/sae_2017_bases_statistiques_formats_sas_csv_7z/"
SAE_2022 = read_csv_from_7z_url(url_SAE_2017)
print(SAE_2022.head(25))

SAE_2022.size()

url = "https://drees2-sgsocialgouv.opendatasoft.com/api/datasets/1.0/708_bases-statistiques-sae/attachments/sae_2023_bases_statistiques_formats_sas_csv_7z/"
```
